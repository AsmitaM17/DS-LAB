```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("myfile.xlsx")
df= df.dropna()
df.plot(x = 'Month', y = 'Sales', kind = 'line', marker ='o', title = 'Monthly salary', color ='b')
plt.ylabel('Amount')
plt.show()

df.plot(x = 'Month', y = 'Expenses', kind = 'bar', title = 'Monthly expenses', color ='g')
plt.ylabel('Amount')
plt.show()
print(df)
```

```python
import matplotlib.pyplot as plt
fruits = ['Orange', 'Banana', 'Mango', 'Apple']
quantities = [20, 25, 25, 30]
colors = ['r', 'b', 'g', 'y', 'k']
plt.pie(quantities, labels=fruits, colors = colors, autopct='%1.1f%%', startangle=90)
plt.title('Fruit Distribution')
plt.axis('equal') #equal aspect ratio makes pie drawn in circle
plt.show()
```

```python
import matplotlib.pyplot as plt
X = [1, 2, 3, 4]
Y = [10, 20, 50, 40]

P = [3, 6, 5, 7]
Q = [5, 15, 25, 35]

plt.scatter(X, Y, color = 'b', marker = 'o', label = 'Dataset 1')
plt.scatter(P, Q, color = 'r', marker = '^', label = 'Dataset 2')

plt.xlabel('X-axis Values')
plt.ylabel('Y-axis Values')
plt.title('Comparison of Two Datasets')
plt.legend()
plt.grid(True, linestyle = '--')
plt.savefig('scatter-comparison.jpg', dpi = 100)
plt.show()
```
